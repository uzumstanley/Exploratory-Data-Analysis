### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('/content/drive/MyDrive/BCCG/DATA/price_data (1).csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

```python
import pandas as pd
# Merge the difference with the original dataframe
df = pd.merge(df, diff, on='id', how='left')

# Calculate the difference between activation date and renewal date
df['diff_activation_renewal'] = (df['date_renewal'] - df['date_activ']).dt.days

# Calculate the difference between modification date and renewal date
df['diff_modif_renewal'] = (df['date_renewal'] - df['date_modif_prod']).dt.days

# Calculate the difference between end date and renewal date
df['diff_end_renewal'] = (df['date_end'] - df['date_renewal']).dt.days


# You can explore more features like:
# - Average consumption per month
# - Consumption trends (increasing/decreasing)
# - Ratio of peak to off-peak consumption
# - Number of days since the last consumption update
# - Interaction terms between existing features

# Example of calculating average consumption
# Assuming you have a 'consommation_mwh' column
# df['avg_monthly_consumption'] = df['consommation_mwh'] / ((df['date_end'] - df['date_activ']).dt.days / 30)


# Drop unnecessary columns
df = df.drop(['date_activ', 'date_end', 'date_modif_prod', 'date_renewal'], axis=1)

# Print the new dataframe
df.head()
```
